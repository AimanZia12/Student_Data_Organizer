```python
print("\n------Welcome to the Student Data Organizer--------")
print("\n------This Student Data Organizer Program is used to collect and store information of different students.------")
students = []
while True:
    print("\n-----Choose any one Option------")
    print("1.Add Student")
    print("2.Display All Students")
    print("3.Update Student Information")
    print("4.Delete Student")
    print("5.Display Subjects Offered")
    print("6.Exit")

    choice_of_user = input("Enter your choice:")
    if choice_of_user == '1':
        print("\n-----User wanted to add student detail-----")
    
        number_of_students = int(input("number of students:"))
        for i in range(number_of_students):
            Student_Id = int(input("student id:"))
            Name_of_Student = input("student name:")
            Age_of_Student = int(input("student age:"))
            Grade_of_Student = input("student grade:")
            Date_of_Birth = input("student date of birth(YYYY-MM-DD):")
            Information_of_Student = (Student_Id, Date_of_Birth)
            Subjects = set(input("student subject:").split(","))

            Details_of_Student = { "Name": Name_of_Student, "Student_Id and DOB": Information_of_Student, "Age": Age_of_Student, "Grade": Grade_of_Student, "Subjects": Subjects}
            students.append(Details_of_Student)
            
        print("\n Student Details Added Successfully!")

    elif choice_of_user == '2':
        print("\n-----User wanted to Display All Studets-----")
        for student in students:
            print(f" ID: {student['Student_Id and DOB'][0]}, Student_Name : {student['Name']}, Age: {student['Age']}, Grade: {student['Grade']}, DOB : {student['Student_Id and DOB'][1]}, Subjects : {', '.join(student['Subjects'])}")

    elif choice_of_user == '3':
        print("\n-------User wanted to Update Student Detail-----")
        Student_Detail_to_Update = int(input("Enter the Student Id which you want to update:"))
        for student in students:
            if student['Student_Id and DOB'][0] == Student_Detail_to_Update:
                student_DOB = list(student['Student_Id and DOB'])
                student['Name'] = input("New_Name:")
                student_DOB[1] = input("New_DOB:")
                student['Age'] = int(input("New_Age:"))
                student['Grade'] = input("New_Grade:")
                pass
        for student in students:
            print(f" ID: {student['Student_Id and DOB'][0]}, Student_Name : {student['Name']}, Age: {student['Age']}, Grade: {student['Grade']}, DOB : {student['Student_Id and DOB'][1]}, Subjects : {', '.join(student['Subjects'])}")
   

    elif choice_of_user == '4':
        print("\n-----User wanted to Delete a student details-----")

        Student_Detail_to_Delete = int(input("Enter the Student Id which you want to delete:"))
        for i in range(len(students)):
            if students[i]['Student_Id and DOB'][0] == Student_Detail_to_Delete:
                del students[i]
                continue
        for student in students:
            print(f" ID: {student['Student_Id and DOB'][0]}\n Student_Name : {student['Name']}\n Age: {student['Age']}\n Grade: {student['Grade']}\n DOB : {student['Student_Id and DOB'][1]}\n Subjects : {', '.join(student['Subjects'])}")

    elif choice_of_user == '5' :
         print("\n-----User wanted to Display all Subjects Offered-----")
         Subjects_Offered = ["Maths", "Statistics", "Data Science", "Marketing", "Finance", "Buisness"]
         print(f" Subjects Offered to Students : {', '.join(Subjects_Offered)}")

    else: 
        print("\n-----User wanted to exit the program-----")
        print("\n-----Thank you, for using Student Data Collector for collecting student data------")
        print("\n-----Exiting from Program------")
        break
        
```
